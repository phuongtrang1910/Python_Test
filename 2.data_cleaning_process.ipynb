## 2. Detect & Remove Outliers Using IQR

```python
df.describe()
```

```python
def detect_outliers_iqr(df, column):
    """
    Detect outliers using the IQR method.
    Outliers = values < Q1 - 1.5*IQR  or > Q3 + 1.5*IQR
    """
    if column not in df.columns:
        raise KeyError(f"Column '{column}' not found in DataFrame.")

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = max(Q1 - 1.5 * IQR, 1)
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

outliers, low, high = detect_outliers_iqr(df, "Quantity")

print("Lower bound:", low)
print("Upper bound:", high)
print("Số outlier:", len(outliers))

```

```python
df = df.drop(outliers.index)
```

# Question 4: Create Recency Column

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

reference_date = df["InvoiceDate"].max() + pd.Timedelta(days=1)

recency = df.groupby("CustomerID").agg({
    "InvoiceDate": lambda x: (reference_date - x.max()).days,  # Recency
})

recency.columns = ["R"]

# ---------------------------
# Min-Max Scaling
# ---------------------------
scaler = MinMaxScaler()
recency_scaled = pd.DataFrame(
    scaler.fit_transform(recency),
    columns=["R_scaled"],
    index=recency.index
)

# Combine original + scaled
recency_final = pd.concat([recency, recency_scaled], axis=1)

print(recency_final)
```

# Question 5: Create Frequency Column

```python
frequency = df.groupby("CustomerID").agg({
    "Invoice": "nunique"   # Count unique invoices per customer
})
frequency.columns = ["F"]

# Min–Max Scale Frequency
frequency_scaled = pd.DataFrame(
    scaler.fit_transform(frequency),
    columns=["F_scaled"],
    index=frequency.index
)

frequency_final = pd.concat([frequency, frequency_scaled], axis=1)

print(frequency_final)
```

# Question 6: Create Monetary Column

```python
df["TotalPrice"] = df["Quantity"] * df["Price"]
```

```python
monetary = df.groupby("CustomerID").agg({
    "TotalPrice": "sum"      # Sum of total revenue
})
monetary.columns = ["M"]

# Min–Max Scale Monetary
monetary_scaled = pd.DataFrame(
    scaler.fit_transform(monetary),
    columns=["M_scaled"],
    index=monetary.index
)

monetary_final = pd.concat([monetary, monetary_scaled], axis=1)
print(monetary_final)
```

# Question 8: Calculate Weighted RFM Score (Apply weights 0.3, 0.5, 0.2 (invert Recency))

```python
# Invert R_Scaled
recency_final["R_scaled"] = 1 - recency_final["R_scaled"]
```

```python
# Merge Recency + Frequency + Monetary
rfm_final = recency_final.merge(
    frequency_final,
    on="CustomerID",
    how="inner"
)

rfm_final = rfm_final.merge(
    monetary_final,
    on="CustomerID",
    how="inner"
)

print(rfm_final)
```

```python
rfm_final[ rfm_final["M_scaled"] == 1 ]
```

```python
rfm_final['RFM_Score'] = rfm_final['R_scaled'] * 0.3 + rfm_final['F_scaled'] * 0.5 + rfm_final['M_scaled'] * 0.2
rfm_final
```
